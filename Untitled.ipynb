```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger vos données dans un DataFrame
data = pd.read_csv('Mesure_horaire_(30j)_Region_Occitanie_Polluants_Reglementaires.csv') 

# Compter le nombre de stations par typologie
stations_par_typologie = data['typologie'].value_counts()
```

```python
# Créer un diagramme à barres pour représenter la répartition des stations par typologie
plt.figure(figsize=(8, 6))
stations_par_typologie.plot(kind='bar', color='skyblue')
plt.title('Répartition des stations par typologie')
plt.xlabel('Typologie')
plt.ylabel('Nombre de stations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()(plt.savefig('Répartition des stations par typologie.csv'))
# Supposons que votre graphe est stocké dans une variable "fig"
plt.savefig('C:\\Users\\SCD UM\\Desktop\\Nouveau dossier\\Répartition des stations par typologie.png')  # Enregistrement du graphe dans un dossier spécifique
data.to_csv('Repartition_des_stations_par_typologie.csv', index=False)
```
